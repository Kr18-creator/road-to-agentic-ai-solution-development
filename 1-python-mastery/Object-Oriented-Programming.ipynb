# Object Oriented Programming in Python

- Python is a mixed language, supporting:
    - Procedural Programming
    - Functional Programming
    - Object-Oriented Programming
- Object-Oriented Programming is a programming paradigm and is based on *objects* and *classes*. It is the process of grouping data and the operations on that data together, structurally.
- Objects
    - Contain both data and behaviours, called *attributes* and *methods*
    - Emulate real-world entities or systems
- Classes
    - Provide a blueprint for creating objects, called *instances*.
    - Define the *type* of instances created from class.
- Guiding Principles of OOP:
    - Encapsulation
    - Inheritance
    - Abstraction
    - Polymorphism
- Encapsulation
    - Bundles data and behaviors in a single unit, an object.
    - The object will have some data it needs and it will be modified only by methods exposed by object. 
    - Promotes Integrity and Modularity.
- Inheritance
    - Create Class hierarchies.
    - Promotes code reuse and reduce duplication.
- Abstraction
    - Hide Implementation details to simplify object interactions.
    - Focus on what, not how.
- Polymorphism
    - Treat objects of different types as a shared base type.
    - Code to interfaces, not types.

- Class Attributes

    - Shared by all instances of the class
- Instance Attributes

## Inheritance in Python

Inheritance
- Lets one class acquire the attributes and methods of another.
- Defines the newly created class as child class, or subclass.
- Defines the original class as the parent, or base class.
- Enables us to extend or repurpose existing class.

With Inheritance, attributes and methods from the parent class can be:
- Inherited directly
  - Parent behavior or data is used as-is
- Overriden
  - Parent behavior or data is replaced.
- Extended
    - Parent behavior or data is augmented or modified.

```python
# Inherited directly
class Parent:
    hair_color = "brown"

class Child(Parent):
    pass
```

```python
child = Child()
child.hair_color
```

- The Parent class is passed in the child class definition: `Child(Parent)`
- `Child` inherits all of the attributes and methods in `Parent`
- The `.hair_color` attribute on `Parent` and `Child` instances is `"brown"`

```python
# Overriden
class Parent:
    hair_color = "brown"

class Child(Parent):
    hair_color = "purple"
```

- `Child` overrides the `.hair_color` attribute defined in `Parent`.
- The `.hair_color` attribute on only `Child` instances in `"purple"`

### Extension

- `super()` lets a class parent methods and access parent attributes.
- `super().__init__()` calls the `.__init__()` of the `Parent` class.

```python
class Parent:
    def __init__(self):
        self.speaks = ["Hindi"]

class Child(Parent):
    def __init__(self):
        super().__init__() # already have speaks attribute
        self.speaks.append("Kannada")
```

```python
class ScienceVessel(Starship):
    ship_type = "Science Vessel"

    def __init__(self, name, crew_size, labs):
        super().__init__(name, crew_size)
        self.labs = labs

    def __str__(self):
        base_str = super().__str__()
        return f"{base_str} | Research labs: {self.labs}"

    def survey(self):
        return f"{self.name} surveys the sector"
```

#### Properties and Descriptors

- The `@property` decorator allows us to create a method that is accessed like an attribute.
- It means we can do calculation to return a value.
- The `@property` decorator is for accessing a value.
- We can also set a value by defining a companion method.
- Decorate the method using the property name and `@.setter`
- Allows us to perform side-effects when a value is set:
    - Perform calculations
    - Caching
    - Error Checks
- Common Pattern:
    - Argument to `.__init__()` passes in initial value.
    - Value is assigned to private attribute.
    - `@property` method with the same name uses the private attribute.
    - `@.setter` method with the same name sets the private attribute with error checking.

```python
class Person:
    def __init__(self, first, last):
        self.first = first
        self.last = last

    @property
    def full_name(self):
        return f"{self.first} {self.last}"
```

```python
himanshu = Person("Himanshu", "Shandilya")
himanshu.first
```

```python
himanshu.full_name
```

```python
# Common Pattern
class Circle:
    def __init__(self, radius):
        self._radius = radius

    @property
    def radius(self):
        return self._radius

    @radius.setter
    def radius(self, value):
        if not isinstance(value, int | float) or value <= 0:
            raise ValueError("Radius must be a positive number")
        self._radius = value
```

```python
small = Circle(3)
small.radius
```

```python
small = Circle('x')
small.radius
```

```python
small.radius = 4
small.radius
```

```python
small._radius
```

```python
small._radius = 5
```

```python
small._radius
```

```python
small.radius = 'x'
```

### Methods, Again...

- There are 3 types of methods in Python:
    - Instance Methods
         - It requires an instantiated object.
         - Typically used to perform operation on associated data.
         - First argument is always an instance of the object(By convention, `self`)
    - Class methods
        - Methods that operates on the class itself.
        - Created using the `@classmethod` decorator.
        - First argument is always a reference to the class(By convention, `cls`)
        - Used for Manipulating data common across all instances, Factories.
    - Static Methods
        - It requires neither a class not an instance.
        - Created using the `@staticmethod` decorator.
        - No required arguments.
        - Typically used for:
            - Grouping data-less functions together.
            - Very seldom needed in Python and can always be achieved with a classmethod.

```python
class Vehicle:
    @classmethod
    def water_vehicle(cls, name, dimensions):
        vehicle = Vehicle()
        vehicle.name = name
        vehicle.dimensions = dimensions
        vehicle.floats = True
        vehicle.num_wheels = 0
        return vehicle

    @classmethod
    def road_vehicle(cls, name, dimensions, num_wheels):
        vehicle = Vehicle()
        vehicle.name = name
        vehicle.dimensions = dimensions
        vehicle.floats = False
        vehicle.num_wheels = num_wheels
        return vehicle

    def volume(self):
        return self.dimensions[0] * self.dimensions[1] * self.dimensions[2]

    @staticmethod
    def all_float(*vehicles):
        for vehicle in vehicles:
            if not vehicle.floats:
                return False
        return True
```

### Putting it all together

```python
# Employee
from datetime import datetime

class Employee:
    company = "MResult Technologies"

    def __init__(self, name, birth_date):
        self.name = name
        self.birth_date = birth_date

    @property
    def birth_date(self):
        return self._birth_date

    @birth_date.setter
    def birth_date(self, value):
        self._birth_date = datetime.fromisoformat(value)

    def compute_age(self):
        today = datetime.today()
        age = today.year - self.birth_date.year
        birthday = datetime(today.year, self.birth_date.month, self.birth_date.day)

        if today < birthday:
            return age - 1
        return age

    @classmethod
    def from_dict(cls, data):
        return cls(**data)

    def __str__(self):
        return f"{self.name} is {self.compute_age()} years old"

    def __repr__(self):
        return (
            "Employee("
            f"""name="{self.name}", """
            f"""birth_date="{self.birth_date.strftime('%Y-%m-%d')}")"""
        )
```

## Python Class Constructors

### Instantiation Process

1. Create a New Instance of the Target Class.
    - Instance Creation:
        - Special Method `.__new__()`
        - Creates and Returns a New Empty Object.
    - Instance Initialization
        - Special method `.__init__()`
        - New Object Passed as `self`
        - Sets Required Instance to a Valid State.
2. Initialize the New Instance with an Appropriate Initial State.

```python
class Point:
    def __new__(cls, *args, **kwargs):
        print("1. Create a new instance instance of Point.")
        return super().__new__(cls)

    def __init__(self, x, y):
        print("2. Initialize the new instance of Point.")
        self.x = x
        self.y = y

    def __repr__(self) -> str:
        return f"{type(self).__name__}(x={self.x}, y={self.y})"
```

```python
point = Point(21, 42)
```
