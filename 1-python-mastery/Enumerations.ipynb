# Enumerations

- How to deal with multiple related constants?
```
RED = 1
GREEN = 2
BLUE = 3

COLORS = (RED, GREEN, BLUE)

pixel_color = {RED: 255, BLUE: 0, GREEN: 200}
pixel_color[RED] -> 255

RED in COLORS -> True
but...
1 in COLORS -> True

is 1 really a color?
```

What we want?

- AN immutable collection of relatable constant members.
    - have unique names (that may have meaning)
    - have an associated constant value
    - unique associated values
    - Operations such as `RED` * 2 or `RED` < `GREEN` are not even allowed
    - Supports enumerating members by name
    - lookup member by name
    - lookup member by value

- How about using a class?
    ```
    class Colors:
        RED = 1
        GREEN = 2
        BLUE = 3
    ```
- How do we look up a member based on its value?
- How do we iterate the members?
- We can but is order preserved?
- Still cannot guarantee uniqueness of value
